1. Three Different Methods for Creating Identical 2D Arrays in NumPy


```python
import numpy as np

# Method 1: Using np.array
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Method 1 Output:\n", array1)

# Method 2: Using np.zeros with manual assignment
array2 = np.zeros((2, 3))
array2[0, :] = [1, 2, 3]
array2[1, :] = [4, 5, 6]
print("Method 2 Output:\n", array2)

# Method 3: Using np.full
array3 = np.full((2, 3), fill_value=[[1, 2, 3], [4, 5, 6]])
print("Method 3 Output:\n", array3)
```

3.1 Explain np.array, np.asarray, and np.asanyarray

np.array: Always creates a new array, even if the input is already an ndarray. It allows you to set properties like dtype.

np.asarray: Converts the input to an ndarray but doesn't copy if the input is already an ndarray with the same dtype.

np.asanyarray: Similar to np.asarray, but passes through subclasses of ndarray. For example, if the input is a matrix, the output will also be a matrix.

3.2 Difference Between Deep Copy and Shallow Copy

Shallow Copy: A shallow copy creates a new object, but does not create copies of nested objects. Changes to nested objects in the original will reflect in the copy.

Deep Copy: A deep copy creates a new object and recursively copies all objects found within the original, ensuring that changes in the original nested objects do not reflect in the copy.

```python
#14.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a date range
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Generate random Temperature and Humidity data
np.random.seed(0)
temperature = np.random.uniform(low=15, high=35, size=30)  # Temperatures between 15°C and 35°C
humidity = np.random.uniform(low=30, high=90, size=30)     # Humidity between 30% and 90%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df.index, df['Temperature'], color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

# Plot Humidity
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df.index, df['Humidity'], color=color, label='Humidity')
ax2.tick_params(axis='y', labelcolor=color)

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show the plot
fig.tight_layout()
plt.show()
```

```python
#15 and 16
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram')

# Calculate the PDF
mu, sigma = 0, 1  # Mean and standard deviation
pdf = norm.pdf(bins, mu, sigma)

# Plot PDF
plt.plot(bins, pdf, 'r--', linewidth=2, label='PDF')

# Label axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set title
plt.title('Histogram with PDF Overlay')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
#17.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.randn(200)
y = np.random.randn(200)

# Determine quadrants
conditions = [
    (x >= 0) & (y >= 0),
    (x < 0) & (y >= 0),
    (x < 0) & (y < 0),
    (x >= 0) & (y < 0)
]
quadrant_labels = ['Quadrant I', 'Quadrant II', 'Quadrant III', 'Quadrant IV']
quadrants = np.select(conditions, quadrant_labels)

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrants})

# Create the scatter plot
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set2', s=100)

# Draw axes lines
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Show legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
#19.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# For Jupyter notebooks
output_notebook()

# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

# Create a DataFrame
df = pd.DataFrame({'Categories': categories, 'Values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values, color=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1', '#a6bddb']))

# Create the figure
p = figure(x_range=categories, height=400, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='categories', top='values', width=0.9, color='color', legend_field='categories', source=source)

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values"),
]
p.add_tools(hover)

# Customize
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
show(p)
```
